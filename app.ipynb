```python
from flask import Flask, request, jsonify
import joblib  # Assuming models are saved as .joblib
import numpy as np

app = Flask(__name__)

# Load your trained models
dnn_model = joblib.load('dnn_model.joblib')
adaboost_model = joblib.load('adaboost_model.joblib')

@app.route('/')
def home():
    return "Welcome to the Parking Prediction Model API"

@app.route('/predict', methods=['POST'])
def predict():
    data = request.json['input']  # Get data from request
    prediction_dnn = dnn_model.predict([data])
    prediction_adaboost = adaboost_model.predict([data])
    
    return jsonify({
        'dnn_prediction': int(prediction_dnn[0]),
        'adaboost_prediction': int(prediction_adaboost[0])
    })

if __name__ == '__main__':
    app.run(debug=True)
```
